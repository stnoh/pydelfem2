# Introduction to the PyDelFEM2  


## Importing the Module


```python
import PyDelFEM2 as dfm2
```

## Importing 3D Mesh

The Delfem2 has a mesh class called `PyDelFEM.Mesh`. The `Mesh` can represent 2D or 3D mesh with a single element type. 

```python
msh = dfm2.Mesh()
msh.read("../test_inputs/bunny_2k.ply")
print("the list of positions is a numpy array: ",type(msh.np_pos))
print("the list of elements is a numpy array: ",type(msh.np_elm))
print(msh.elem_type) # element type
```

## Draw 3D Mesh

you can draw a 3D mesh using image viewer or 3D window. In this Jupyter notebook page, we only demonstrate Drawing the mesh using the image viewer such as the `pyplot`.

The function 'imgDraw3d' outputs byte RGB array of the image. We can directory draw the image with the `pyplot` as

```python
%matplotlib inline
import matplotlib.pyplot as plt
import PyDelFEM2.gl.glfw

msh.normalize()
assert msh is not None
img = dfm2.gl.glfw.imgDraw3d([msh],winsize=(400,300))

plt.imshow(img)
plt.show()
```
